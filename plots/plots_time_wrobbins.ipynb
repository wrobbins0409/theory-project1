```python
import pandas as pd
import matplotlib.pyplot as plt

# load in times
df = pd.read_csv('output_times.csv')

# Display the DataFrame
display(df)
```

```python
# Group the rows into chunks of 10 and calculate the mean for each chunk
df_avg = df.groupby(df.index // 10).mean()

# Plot the averaged data with two lines
plt.figure(figsize=(10, 6))
plt.plot(df_avg['TwoSAT'], label='TwoSAT', marker='o')
plt.plot(df_avg['DumbSAT'], label='DumbSAT', marker='x')

# Add labels and a title
plt.xlabel('Grouped WFFs (every 10)')
plt.ylabel('Average Time')
plt.title('Average Times for TwoSAT and DumbSAT (Grouped by 10 WFFs)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
